# <span style="color:deepskyblue"> **Exploratory Data Analysis**

## <span style="color:deepskyblue"> **Categorical data**

It's great that you're performing Exploratory Data Analysis (EDA) on your dataset as part of your journey to become a Data Analyst. Visualizing categorical columns is essential for gaining insights into the distribution and relationships within your data. Here are some common visualization techniques you can use for categorical data:

1. **Bar Charts:**
   - Bar charts are one of the most straightforward ways to visualize categorical data. You can create a bar chart for each categorical column to see the distribution of categories.
   - For example, you can create a bar chart for the "Ship Mode" column to visualize the distribution of different shipping modes.

2. **Pie Charts:**
   - Pie charts are useful when you want to visualize the proportions of categories within a single categorical column.
   - For example, you can create a pie chart to visualize the proportions of different segments (e.g., Consumer, Corporate, Home Office) in the "Segment" column.

3. **Countplots (Seaborn):**
   - Seaborn's countplot is a specialized bar plot for showing the count of observations in each category.
   - It can be very effective for visualizing the frequency of each category within a column.

4. **Stacked Bar Charts:**
   - If you want to visualize the distribution of categories across two or more categorical columns, you can create stacked bar charts.
   - For instance, you can create a stacked bar chart to show how different shipping modes are distributed within each segment.

5. **Heatmaps:**
   - Heatmaps can be used to visualize the relationships between two categorical variables. You can create a heatmap to see how different categories in two columns interact.
   - For example, you can create a heatmap to see if there are any patterns in the distribution of "Region" and "Category."

6. **Mosaic Plots:**
   - Mosaic plots are another way to visualize the relationships between two or more categorical variables. They show the relative frequencies of categories and can help identify associations.
   - You can use a mosaic plot to explore the relationships between "Region," "Category," and "Sub-Category."

7. **Word Clouds:**
   - If your categorical columns contain text data (e.g., product names or city names), you can create word clouds to visualize the most frequent words or phrases.
   - This can be useful for columns like "City" or "Sub-Category" if they contain textual information.

8. **Violin Plots or Box Plots:**
   - Violin plots or box plots can be used to visualize the distribution of a numerical variable within each category of a categorical column.
   - For instance, you can create a violin plot to see the distribution of order counts within each "Category."

# Stacked Bar Chart - Ship Mode and Segment

```python
# Use crosstab to create a contingency table for Ship Mode and Segment
cross_tab = pd.crosstab(superstore_df['Segment'], superstore_df['Ship Mode'])


# Create a stacked bar chart
plt.figure(figsize=(10, 6))
cross_tab.plot(kind='bar', stacked=True, colormap='viridis')


plt.title('Stacked Bar Chart of Ship Mode by Segment')
plt.xlabel('Segment')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.legend(title='Ship Mode', loc='upper right')


plt.show()
print(cross_tab)
```

In this code:

1. We use `pd.crosstab` from the pandas library to create a contingency table that shows the count of each combination of "Segment" and "Ship Mode."

2. We create a stacked bar chart by setting `stacked=True` when calling the `plot()` method on the `cross_tab` DataFrame.






# Heatmap - Region vs Category

```python
# Create a cross-tabulation of Region and Category
cross_tab = pd.crosstab(superstore_df['Region'], superstore_df['Category'])


# Create a heatmap using Seaborn
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlGnBu')


plt.title('Heatmap of Region vs. Category')
plt.xlabel('Category')
plt.ylabel('Region')



plt.show()
print(cross_tab)
```

In this code:

1. We first use `pd.crosstab` to create a cross-tabulation or contingency table of "Region" and "Category." This table shows the count of each combination of region and category.

2. We create a heatmap using Seaborn's `heatmap()` function. We provide the cross-tabulation as the data, set `annot=True` to display the cell values, use `fmt='d'` to format cell values as integers, and choose a colormap (`cmap`) for the heatmap.

3. We set the title, labels for the x-axis and y-axis using `plt.title()`, `plt.xlabel()`, and `plt.ylabel()`.

4. Finally, we display the heatmap using `plt.show()`.
